# 3. 데이터 분석 및 시각화
데이터에 적합한 그래프를 선택(1개 이상 + 히트맵)하여 시각화하고, 그 의미를 논리적으로 해석한다.

시각화한 내용 마다 해석한 결과 즉, **"이 그래프가 나의 문제 해결(예측/분류)에 어떤 실마리를 주는지"** 반드시 적도록 한다.


| 그래프 종류 | 목적 | 코드(예) |
| :--- | :--- | :--- |
| **박스 플롯**<br>(`sns.boxplot`) | 데이터의 사분위수(Q1, Q2, Q3)와 **이상치를 한눈에 파악**하고 그룹별로 비교할 때 | `sns.boxplot(data=df, x='Group', y='Score')` |
| **바 플롯**<br>(`sns.barplot`) | 각 범주(그룹)별 **평균값의 크기를 막대 길이로 비교**할 때 | `sns.barplot(data=df, x='Group', y='Score')` |
| **카운트 플롯**<br>(`sns.countplot`) | 범주형 변수의 **데이터 개수(빈도수)를 그룹별로 세어** 막대로 표현할 때 | `sns.countplot(data=df, x='Group')` |
| **히스토그램**<br>(`sns.histplot`) | 연속형 변수의 구간별 빈도를 나타내어 전체적인 **데이터의 분포 모양을 확인**할 때 | `sns.histplot(data=df, x='Age', bins=20, kde=True)` |
| **커널 밀도 그래프**<br>(`sns.kdeplot`) | 히스토그램을 부드러운 곡선으로 정형화하여 **확률 밀도 분포를 시각화**할 때 | `sns.kdeplot(data=df, x='Age', fill=True)` |
| **산점도**<br>(`sns.scatterplot`) | 두 수치형 변수 간의 **상관관계나 데이터의 분포 경향**을 점으로 나타낼 때 | `sns.scatterplot(data=df, x='Height', y='Weight', hue='Gender')` |
| **선 그래프**<br>(`sns.lineplot`) | 주로 '시간' 흐름에 따른 데이터의 **연속적인 변화 추세(시계열)** 를 볼 때 | `sns.lineplot(data=df, x='Date', y='Sales')` |
| **회귀선 플롯**<br>(`sns.regplot`) | 산점도 위에 두 변수의 관계를 가장 잘 설명하는 **직선(회귀선)을 함께 그릴 때** | `sns.regplot(data=df, x='StudyTime', y='Score')` |
| **히트맵**<br>(`sns.heatmap`) | 변수 간의 상관계수 등 **2차원 배열 데이터를 색상의 진하기로 표현**할 때 | `sns.heatmap(df.corr(), annot=True, cmap='coolwarm')` |
| **페어 플롯**<br>(`sns.pairplot`) | 데이터프레임 내 **모든 수치형 변수 쌍의 산점도와 분포**를 격자로 한 번에 그릴 때 | `sns.pairplot(df, hue='Gender', corner=True)` |
